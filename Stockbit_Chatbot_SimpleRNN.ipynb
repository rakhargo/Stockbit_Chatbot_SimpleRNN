### Import data

```python
import pandas as pd

df = pd.read_csv('StockbitFAQ.csv')

df
```

```python
text_column = 'content'
title_column = 'title'

# Make a custom text splitter function
def custom_text_splitter(text, chunk_size, chunk_overlap):
    start = 0
    chunks = []
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - chunk_overlap  # Move start back by chunk_overlap to create overlap
    return chunks

split_data = []

for index, row in df.iterrows():
    title = row[title_column]
    content = row[text_column]
    # Ensure content is a string
    if not isinstance(content, str):
        content = str(content)
    # Set chunk size to 500 char, 50 char overlap    
    chunks = custom_text_splitter(content, 500, 50)
    for i, chunk in enumerate(chunks):
        if i == 0:
            # Concatenate title with the first chunk
            # This is to improve data retrieval accuracy
            chunk = title + "\n\n" + chunk
        split_data.append({'title': title, 'content': chunk})

# Create a new DataFrame with the split content
split_df = pd.DataFrame(split_data)

# Display the split DataFrame
print(split_df.head())
```

### Proses data

```python
import re

split_df['content'] = split_df['content'].apply(lambda x: re.sub(r'http\S+|www\S+|https\S+', '', x)) # url
split_df['content'] = split_df['content'].apply(lambda x: re.sub(r'\(.*?\)|\{.*?\}|\[.*?\]', '', x)) # hapus dalam kurung
split_df['content'] = split_df['content'].apply(lambda x: re.sub(r'[^A-Za-z0-9\s.,]', '', x)) # menghapus selain huruf angka , .
split_df['content'] = split_df['content'].apply(lambda x: re.sub(r'colorabb', '', x)) # menghapus 'colorabb'

display(split_df)
```

```python
titles = split_df['title']
contents = split_df['content']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(titles + contents)
vocab_size = len(tokenizer.word_index) + 1

title_sequences = tokenizer.texts_to_sequences(titles)
content_sequences = tokenizer.texts_to_sequences(contents)

max_length = max(max(len(seq) for seq in title_sequences), max(len(seq) for seq in content_sequences))
title_sequences_padded = pad_sequences(title_sequences, maxlen=max_length, padding='post')
content_sequences_padded = pad_sequences(content_sequences, maxlen=max_length, padding='post')
```

```python
from sklearn.model_selection import train_test_split

X_train = title_sequences_padded
Y_train = content_sequences_padded
```

### Pembuatan Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(Embedding(vocab_size, 100))
model.add(SimpleRNN(128, return_sequences=True))
model.add(SimpleRNN(64, return_sequences=True))
model.add(Dense(vocab_size, activation='softmax'))


model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
import numpy as np

X_train = content_sequences_padded[:, :-1]  # Semua kata dalam sequence kecuali yang terakhir
y_train = content_sequences_padded[:, 1:]   # Semua kata dalam sequence kecuali yang pertama (untuk prediksi kata berikutnya)

model.fit(X_train, y_train, epochs=150)
```
